```python
#Import all of the 10 models
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
```

```python
#Model building
symptoms_disease_models = {
    "Random Forest" : RandomForestClassifier(),
    "Logistic Regression" : LogisticRegression(),
    "K-Nearest Neighbor" : KNeighborsClassifier(),
    "Support Vector Machine" : SVC(probability=True),
    "Extreme Gradient Boosting" : XGBClassifier(),
    "Cat Boosting" : CatBoostClassifier(),
    "Light GBM" : LGBMClassifier(),
    "Multinomial Naive Bayes" : MultinomialNB(),
    "Gaussian Naive Bayes" : GaussianNB(),
    "Bernoulli Naive Bayes" : BernoulliNB()
}
```

```python
for name_of_model,model in symptoms_disease_models.items():
  print(model)
```

```python
#Encoding
from sklearn.preprocessing import LabelEncoder
disease_encoder = LabelEncoder()

#Disease encoding
disease_columns = data_set.Disease
encoded_disease = disease_encoder.fit_transform(disease_columns)


#Symptoms encoding
symptom_encoder = LabelEncoder()
symptom_columns = data_set.drop('Disease',axis=1)
twod_np_symptoms = symptom_columns.values
oned_symptoms = twod_np_symptoms.flatten()
symptom_encoder.fit(oned_symptoms)
encoded_symptoms = symptom_columns.apply(symptom_encoder.transform)


#Join back the disease and symptoms columns
encoded_dataset = encoded_symptoms
encoded_dataset.insert(0,'Disease',encoded_disease)
```

```python
print(encoded_dataset)
```

```python
#Splitting
input_symptoms = encoded_dataset.drop('Disease',axis=1)
output_disease = encoded_dataset.Disease

from sklearn.model_selection import train_test_split
symptom_train,symptom_test,disease_train,disease_test = train_test_split(input_symptoms,output_disease,test_size=0.2)
```

```python
#Training the models
for name_of_model,model in symptoms_disease_models.items():
  model.fit(symptom_train,disease_train)
```

```python
#Prediction
for name_of_model,model in symptoms_disease_models.items():
  prediction = model.predict(symptom_test)
  prediction = disease_encoder.inverse_transform(prediction)
  print(prediction)
```

```python
#Decode the prediction back to categorical data
#prediction = disease
#Predicted diseases have been decoded
#Decode the actual diseases which are disease_test
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
import warnings
warnings.filterwarnings('ignore')
accuracy_list =[ ]
recall_list = [ ]
precision_list = [ ]
name_list = [ ]
fone_list = [ ]
auc_list = [ ]
for name,model in symptoms_disease_models.items():
  name_list.append(name)
  prediction = model.predict(symptom_test)
  accuracy = accuracy_score(disease_test,prediction)
  accuracy_list.append(accuracy)
  precision = precision_score(disease_test,prediction,average='macro')
  precision_list.append(precision)
  recall = recall_score(disease_test,prediction,average='macro')
  recall_list.append(recall)
  fone = f1_score(disease_test,prediction,average='macro')
  fone_list.append(fone)
  auc_prediction = model.predict_proba(symptom_test)
  auc_roc = roc_auc_score(disease_test,auc_prediction,multi_class='ovr',average='macro')
  auc_list.append(auc_roc)


metrics_dictionary = {
    "Name of the model" : name_list,
    "Accuracy score" : accuracy_list,
    "Precision score" : precision_list,
    "Recall score" : recall_list,
    "F1 score" : fone_list,
    "AUC-ROC score" : auc_list

}
metrics_dataframe = pd.DataFrame(metrics_dictionary)
metrics_dataframe = metrics_dataframe.sort_values(by="AUC-ROC score",ascending=False)
print(metrics_dataframe)
```
